```python
n = int(input("enter any number :"))
rev = 0
a = n
while (n>0):
  r = n%10
  n = n//10
  rev = rev * 10 + r
print("revers number is:",rev)
if(rev==a):
  print("it is a palindrom")
else:
  print("it is not a palindrom")
```
