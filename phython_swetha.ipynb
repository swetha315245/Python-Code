```python
pcost   = eval(input("enter product cost :"))
qty     = eval(input("enter quantity :"))
billamt = pcost * qty

if(billamt > 5000):
  dis = 20
  disamt = billamt * dis /100
  famt = billamt - disamt
  print("bill amount :",billamt)
  print("discount amount :",disamt)
  print("final amount :",famt)
else:
  print("no discount")
  print("final bill amount",billamt)
```
